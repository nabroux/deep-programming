```python
import random, time

class Character():
    def __init__(self, name):
        self.name = name
        self.earth = random.randint(1,1000)
        self.atk = random.randint(100,1000)
        self.defe = random.randint(100,1000)
        self.hp = 2000
        
        abi = self.atk + self.defe
        if abi <= 500:
            self.occu = '路人'
        elif abi <= 800:
            self.occu = '打手'
        elif abi <= 1200:
            self.occu = '強者'
        elif abi <= 1800:
            self.occu = '天才'
        elif abi <= 2000:
            self.occu = '王者'
            
        if self.earth == 1:
            self.occu = '上班族'
            
    def show_info(self):
        print('\n【來自{}號地球的{}.....攻擊力為{}，防禦力為{}，是個{}。】'.format(self.earth, self.name,
                                                         self.atk, self.defe, self.occu))
        
class Fight():
    def __init__(self, warriors1, warriors2):
        self.warriors1 = warriors1
        self.warriors2 = warriors2
        
        self.warriorsList = [self.warriors1, self.warriors2]
        
    def _load(self):
        print('戰場載入中', end='')
        for i in range(5):
            time.sleep(0.5)
            print('.', end='')
        print('【戰場載入完成！】')
            
    def fight(self):
        self._load()
        
        print('【戰鬥開始】')    
        while True:
            time.sleep(random.randint(1,3))
            attacker, defender = random.sample(self.warriorsList, 2)
            print('{}對{}發動了攻擊!'.format(attacker.name, defender.name))
            time.sleep(1)
            attackProb = random.random()
            if attackProb < 0.5:                
                minDamage = int(max(1.5*attacker.atk - 0.5*defender.defe, 0))
                maxDamage = int(1.5 * attacker.atk)
                damage = random.randint(minDamage, maxDamage)
                print('對{}造成了{}點傷害'.format(defender.name, damage))
                defender.hp -= damage
            elif attackProb < 0.65:
                minDamage = int(max(2.5*attacker.atk - 0.5*defender.defe, 0))
                maxDamage = int(3.5 * attacker.atk)
                damage = random.randint(minDamage, maxDamage)
                print('{}施展出了會心一擊！'.format(attacker.name))
                print('對{}造成了{}點傷害'.format(defender.name, damage))
                defender.hp -= damage
            else:
                print('{}成功閃躲!'.format(defender.name))

            if random.randint(1,10) == 1:
                time.sleep(1)
                print('{}找到了帶來的道具！'.format(defender.name))
                if random.randint(1,5) == 1:
                    print('{}喝了特級體力藥水！'.format(defender.name))
                    print('{}回復了200點生命值！'.format(defender.name))
                    defender.hp += 1000
                else:
                    print('{}喝了體力藥水！'.format(defender.name))
                    print('{}回復了100點生命值！'.format(defender.name))
                    defender.hp += 500

            if defender.hp <= 0:
                print('{}陣亡！獲勝者是{}'.format(defender.name, attacker.name))
                break
                
            print('【{}生命值:{}】 【{}生命值:{}】\n'.format(self.warriors1.name, self.warriors1.hp, self.warriors2.name, self.warriors2.hp))

```

```python
characters = []
character_name = ''
while len(characters) < 2:
    name_input = input('請輸入角色名稱:')
    if name_input != '':
        character_name = name_input
    character = Character(character_name)
    character.show_info()
    
    print('\n')
    isChosen = input('滿意嗎？(y/n)')
    if isChosen == 'y':
        characters.append(character)
    
f = Fight(characters[0], characters[1])
f.fight()
```
