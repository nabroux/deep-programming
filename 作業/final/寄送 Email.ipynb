# 資管四 105306076 許雲輔 寄送Email (SMTP)

```python
#應用程式 key 已隱藏
import smtplib
from email.mime.text import MIMEText

gmail_user = 'philasofa@gmail.com'
gmail_password = 'xxxxxxxxxxx' # your gmail application key

content = '收到這封信的你，會得到好運'
msg = MIMEText(content)
msg['Subject'] = 'Test'
msg['From'] = gmail_user

receiver = input('收信人: ')
msg['To'] = receiver

server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
print(server.ehlo())
server.login(gmail_user, gmail_password)
server.send_message(msg)
server.quit()

print('\n【Email 成功送出!】')
```
