利用 pytesseract 做超簡易OCR應用，中文有時候效果不佳

```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt

def ocr_demo(fileName, lang):
    img = Image.open(fileName)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    text = pytesseract.image_to_string(img, lang=lang)
    print(text)

print('\n\n【繁體模式】')
ocr_demo('chi_tra.jpg', 'chi_tra')
    
print('\n\n【簡體模式】')
ocr_demo('chi_sim.jpg', 'chi_sim')

print('\n\n【英文模式】')
ocr_demo('eng.png', 'eng')
```
