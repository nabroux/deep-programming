# 105306076 資管四 許雲輔 菱骨塔

```python
from ipywidgets import interact

def showDiamond(size):
    line=''
    structure = list(range(1,int(size)+1))[::2] + list(range(int(size)-2,0,-1))[::2]

    for i in structure:
        line += int((size-i)/2) * ' ' + i * '*' + '\n'

    print(line)

print('菱骨塔要蓋多高？')
interact(showDiamond , size=(1,100,2));
```
