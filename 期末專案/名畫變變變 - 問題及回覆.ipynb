# Q1. 希望解釋一下content weight(0.08)是怎麼求得的？

 

程式碼中的 content weight 指的是「從內容圖提取出的內容之權重」，會影像到成品的結果。

content weight 值越高，代表成品的「內容」部分會越強烈，反之則越不明顯。

相反的，底下的 feature layers w 則代表了各層風格提取的權重，也會直接影響成品結果。

因此，這些權重不太像是求得的，而是經由多次實驗，我們認為效果最好的權重，這個值也不一定要是 0.08 ~
